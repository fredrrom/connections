# Prover vs ground truths for the TPTP v3.3.0 SYN problems

For overall test statistics scroll to bottom.

## Collecting ground truths and other metadata:

```python
from tqdm.notebook import tqdm
import benchmark_prover as bp
import pandas as pd
import os

classical_rows = {}
intuitionistic_rows = {}
cnf_intu_files = "../../conjectures/TPTP-v3.3.0_SYN_icnf"
cnf_files = "../../conjectures/TPTP-v3.3.0_SYN_noeq_cnf"
fof_files = "../../conjectures/TPTP-v3.3.0/Problems/SYN"
print(f'Number of CNF files: {len(os.listdir(cnf_intu_files))}')

for file in tqdm(os.listdir(fof_files), desc="Collecting FOF metadata: "):
    bp.collect_metadata(fof_files, file, classical_rows)
    bp.collect_metadata(fof_files, file, intuitionistic_rows)
```

## Benchmarking pyCoP and leanCoP:

```python
for file in tqdm(os.listdir(cnf_intu_files), desc='pyCoP proving SYN theorems (1s Timeout) : '):
    cmd = ["python3", "leancop.py"]
    bp.run(cmd, 'pyCoP', cnf_intu_files, file, classical_rows, 1)

for file in tqdm(os.listdir(cnf_files), desc='leanCoP proving SYN theorems (1s Timeout) : '):
    cmd = ["./leancop10c.sh " + os.path.join(cnf_files, file)]
    bp.run(cmd, 'leanCoP', cnf_intu_files, file, classical_rows, 1)

classical_df = pd.DataFrame.from_dict(classical_rows, orient='index').reset_index(drop=True).dropna()
display(classical_df)
```

## Benchmarking ipyCoP and ileanCoP:

```python
for file in tqdm(os.listdir(cnf_intu_files), desc='ipyCoP proving SYN theorems (1s Timeout) : '):
    cmd = ["python3", "../../experiments/ileancop.py"]
    bp.run(cmd, 'ipyCoP', cnf_intu_files, file, intuitionistic_rows, 1)

intuitionistic_df = pd.DataFrame.from_dict(intuitionistic_rows, orient='index').reset_index(drop=True).dropna()
ileancop_results = pd.read_csv('results_tptp330_intuit_nov10', sep="\s+", skiprows=137, usecols=[0,6], header=None)

def status(x):
    non = False
    if x[0] == '(':
        non = True
        x = x[1:-1]
    if x[0] == '<' and float(x[1:]) < 1:
        return 'Non-Theorem' if non else 'Theorem'
    if x[0] == 's':
        return 'Stack'
    if x[0] == '>' or float(x) > 1:
        return 'Timeout'
    return 'Non-Theorem' if non else 'Theorem'

def time(x):
    if x[0] == '(':
        x = x[1:-1]
    if x[0].isnumeric() and float(x) < 1:
        return float(x)
    if x[0] == '<' and float(x[1:]) < 1:
        return float(x[1:])
    return 1

ileancop_results['ileanCoP Status (1s)'] = ileancop_results[6].apply(status)
ileancop_results['ileanCoP Time (1s)'] = ileancop_results[6].apply(time)
ileancop_results.rename(columns={0:'File'},inplace=True)
intuitionistic_df = intuitionistic_df.merge(ileancop_results.drop(columns=[6]), how='left', on='File')
```

## Summary of pyCoP performance:

```python
summary = classical_df.groupby(classical_df.File.str[:3])['File'].count().rename_axis('Status/pyCoP').to_frame(name = 'Total')
summary['Theorem/Theorem'] = classical_df.loc[(classical_df['Status'] == 'Theorem') & (classical_df['pyCoP Status (1s)'] == 'Theorem')].groupby(classical_df.File.str[:3])['File'].count()
summary['Theorem/Timeout'] = classical_df.loc[(classical_df['Status'] == 'Theorem') & (classical_df['pyCoP Status (1s)'] == 'Timeout')].groupby(classical_df.File.str[:3])['File'].count()
summary['CounterSatisfiable/Timeout'] = classical_df.loc[(classical_df['Status'] == 'CounterSatisfiable') & (classical_df['pyCoP Status (1s)'] == 'Timeout')].groupby(classical_df.File.str[:3])['File'].count()
summary['CounterSatisfiable/Theorem'] = classical_df.loc[(classical_df['Status'] == 'CounterSatisfiable') & (classical_df['pyCoP Status (1s)'] == 'Theorem')].groupby(classical_df.File.str[:3])['File'].count()
summary = summary.reindex(columns=['Theorem/Theorem','Theorem/Timeout','CounterSatisfiable/Timeout','CounterSatisfiable/Theorem','Total'])
summary = summary.fillna(0).T.astype(int)
display(summary)
```

## Summary of LeanCoP performance

```python
summary = classical_df.groupby(classical_df.File.str[:3])['File'].count().rename_axis('Status/LeanCoP').to_frame(name = 'Total')
summary['Theorem/Theorem'] = classical_df.loc[(classical_df['Status'] == 'Theorem') & (classical_df['leanCoP Status (1s)'] == 'Theorem')].groupby(classical_df.File.str[:3])['File'].count()
summary['Theorem/Timeout'] = classical_df.loc[(classical_df['Status'] == 'Theorem') & (classical_df['leanCoP Status (1s)'] == 'Timeout')].groupby(classical_df.File.str[:3])['File'].count()
summary['CounterSatisfiable/Timeout'] = classical_df.loc[(classical_df['Status'] == 'CounterSatisfiable') & (classical_df['leanCoP Status (1s)'] == 'Timeout')].groupby(classical_df.File.str[:3])['File'].count()
summary['CounterSatisfiable/Theorem'] = classical_df.loc[(classical_df['Status'] == 'CounterSatisfiable') & (classical_df['leanCoP Status (1s)'] == 'Theorem')].groupby(classical_df.File.str[:3])['File'].count()
summary = summary.reindex(columns=['Theorem/Theorem','Theorem/Timeout','CounterSatisfiable/Timeout','CounterSatisfiable/Theorem','Total'])
summary = summary.fillna(0).T.astype(int)
display(summary)
```

## Checking for errors:
- **False positive**: Prover claims formula is a theorem while the file claims the formula is countersatisfiable.

```python
fp_lean = classical_df.loc[(classical_df['Status'] == 'CounterSatisfiable') & (classical_df['leanCoP Status (1s)'] == 'Theorem')]
fp_py = classical_df.loc[(classical_df['Status'] == 'CounterSatisfiable') & (classical_df['pyCoP Status (1s)'] == 'Theorem')]
if len(fp_lean.index) > 0:
    print("\u2718 leanCoP false positives (1s): %d (%.2f%%)" % (len(fp_lean.index), 100*len(fp_lean.index)/len(classical_df.index)))
    display(fp_lean.style.hide_index())
else:
    print('\u2714 leanCoP false positives (1s): 0 (0.00%%)')
if len(fp_py.index) > 0:
    print("\u2718 pyCoP false positives (1s): %d (%.2f%%)" % (len(fp_py.index), 100*len(fp_py.index)/len(classical_df.index)))
    display(fp_py.style.hide_index())
else:
    print('\u2714 pyCoP false positives (1s): 0 (0.00%%)')
```

## Checking number of correct conclusions:
- **True positive**: Both prover and file claims the formula is a theorem.
- **True negative**: Prover cannot prove that the formula is a theorem and the file claims the formula is countersatisfiable.
- **False negative**: Prover cannot prove that the formula is a theorem while the file claims the formula is a theorem.

```python
tp_lean = classical_df.loc[(classical_df['Status'] == 'Theorem') & (classical_df['leanCoP Status (1s)'] == 'Theorem')]
tn_lean = classical_df.loc[(classical_df['Status'] == 'CounterSatisfiable') & (classical_df['leanCoP Status (1s)'] == 'Timeout')]
fn_lean = classical_df.loc[(classical_df['Status'] == 'Theorem') & (classical_df['leanCoP Status (1s)'] == 'Timeout')]
print("Results for leanCoP given 1s:")
print('\u2714 True positives: %d (%.2f%%)' % (len(tp_lean.index), 100*len(tp_lean.index)/len(classical_df.index)))
print('\u2714 True negatives: %d (%.2f%%)' % (len(tn_lean.index), 100*len(tn_lean.index)/len(classical_df.index)))
print('\u2718 False negatives: %d (%.2f%%)' % (len(fn_lean.index), 100*len(fn_lean.index)/len(classical_df.index)))
print('\nTotal successes: %d (%.2f%%)' % (len(tp_lean.index)+len(tn_lean.index), \
                                          100*(len(tp_lean.index)+len(tn_lean.index))/len(classical_df.index)))
print(len(classical_df.index))
print(len(tp_lean.index)+len(tn_lean.index)+len(fp_lean.index)+len(fn_lean.index))
if (len(tp_lean.index)+len(tn_lean.index)+len(fp_lean.index)+len(fn_lean.index)) == len(classical_df.index):
    print('\u2714 Sanity check (PASS)') 
else: 
    print('\u2718 Sanity check (FAIL)')
```

```python
tp_py = classical_df.loc[(classical_df['Status'] == 'Theorem') & (classical_df['pyCoP Status (1s)'] == 'Theorem')]
tn_py = classical_df.loc[(classical_df['Status'] == 'CounterSatisfiable') & (classical_df['pyCoP Status (1s)'] == 'Timeout')]
fn_py = classical_df.loc[(classical_df['Status'] == 'Theorem') & (classical_df['pyCoP Status (1s)'] == 'Timeout')]
print("Results for pyCoP given 1s:")
print('\u2714 True positives: %d (%.2f%%)' % (len(tp_py.index), 100*len(tp_py.index)/len(classical_df.index)))
print('\u2714 True negatives: %d (%.2f%%)' % (len(tn_py.index), 100*len(tn_py.index)/len(classical_df.index)))
print('\u2718 False negatives: %d (%.2f%%)' % (len(fn_py.index), 100*len(fn_py.index)/len(classical_df.index)))
print('\nTotal successes: %d (%.2f%%)' % (len(tp_py.index)+len(tn_py.index), \
                                          100*(len(tp_py.index)+len(tn_py.index))/len(classical_df.index)))
print(len(classical_df.index))
print(len(tp_py.index)+len(tn_py.index)+len(fp_py.index)+len(fn_py.index)) 
if (len(tp_py.index)+len(tn_py.index)+len(fp_py.index)+len(fn_py.index)) == len(classical_df.index):
    print('\u2714 Sanity check (PASS)') 
else: 
    print('\u2718 Sanity check (FAIL)')
```
