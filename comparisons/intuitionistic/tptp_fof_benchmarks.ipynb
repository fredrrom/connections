```python
from tqdm.notebook import tqdm
import benchmark_prover as bp
import subprocess as sp
import pandas as pd
import os

rows = {}

translator_path = 'leancop_trans_v22f/leancop_trans.sh'
leancop_path = 'leancop10f_trace/leancop10.sh'
pycop_path = 'leancop.py'
fof_folders = "../../conjectures/TPTP-v6.4.0/Problems"
fof_filter = lambda filename : '+' in filename and '.p' in filename
for folder in os.listdir(fof_folders):
    folderpath = os.path.join(fof_folders, folder)
    if not os.path.isdir(folderpath):
        continue
    for file in tqdm(os.listdir(folderpath), desc=f"Collecting {folder} metadata: "):
        if fof_filter(file) :
            bp.collect_metadata(folderpath, file, rows)
```

```python
df = pd.DataFrame.from_dict(rows, orient='index').reset_index(drop=True).dropna()
df
```

```python
for folder in os.listdir(fof_folders):
    folderpath = os.path.join(fof_folders, folder)
    if not os.path.isdir(folderpath):
        continue
    for file in tqdm(os.listdir(folderpath), desc=f"pyCoP proving {folder} (1s Timeout) : "):
        if fof_filter(file) :
            filepath = os.path.join(folderpath, file)
            problem = os.path.splitext(file)[0] + '.cnf'
            cmd = [translator_path, filepath, problem]
            bp.run_with_time_limit(cmd, 3)
            #sp.run([translator_path, filepath, problem])
            if not os.path.exists(problem):
                del rows[os.path.splitext(file)[0]]
                continue
            print(problem)
            #bp.run(f'python {pycop_path} {problem}', 'pyCoP', '', file, rows, 1)
            os.remove(problem)
```

```python
for folder in os.listdir(fof_folders):
    folderpath = os.path.join(fof_folders, folder)
    if not os.path.isdir(folderpath):
        continue
    for file in tqdm(os.listdir(folderpath), desc=f"leancop proving {folder} (1s Timeout) : "):
        if not '-' in file and '.p' in file :
            filepath = os.path.join(folderpath, file)
            cmd = [leancop_path + "" + filepath]
            bp.run(cmd, 'leanCoP', '', file, rows, 1)
```
